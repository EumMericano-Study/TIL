# 시각화

* 이해력, 관상용(관심도, 커뮤니케이션) 증대
* **변화**에 관한 그래프

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/lec8_Visualization/broadcast.csv', index_col=0)
df
```

```python
df.plot() # x 축은 인덱스 (일반적으로 시간 값이 적합) 
df.plot(kind='line')  #kind 파라미터 -> 기본값 line
```

```python
df.plot(y='KBS')  # y값 설정 가능
df.plot(y=['KBS', 'JTBC'])  # y값 설정 가능
```

```python
df['KBS']
```

```python
df['KBS'].plot() # pandas Series에도 plot 메서드가 존재
```

### DataFrame의 값들은 모두 숫자여야 한다  문자로 그래프를 그리려면 **오류**가 발생

## 막대 그래프
* **항목**비교 그래프

```python
sport_df = pd.read_csv('data/lec8_Visualization/sports.csv', index_col = 0)
sport_df
```

```python
sport_df.plot() # 선그래프는 이해하기 힘들다 -> 적합한 표를 찾아야 한다
```

```python
sport_df.plot(kind='bar', stacked=True) # 쌓아올리기
```

## 히스토그램
* **범위**를 표시하는 그래프
* 범위를 마음대로 설정 가능

```python
height_df = pd.read_csv('data/lec8_Visualization/body.csv')
height_df.head(5)
```

```python
height_df.plot(kind='hist', y='Height', bins=15) # bins = 범위의 개수 (기본값=10)
```

```python
height_df.plot(kind='hist', y='Height', bins=200) # bins = 범위의 개수
```

### bins가 높다고 무조건 좋은것만은 아니다. 표시하기 전에 주의할것
